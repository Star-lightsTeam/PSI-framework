```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data.csv")
```

```python
import requests  
url = df['商品的图片信息']

filename = '.jpg'   
for i in range (len(url)):
    num=i
    response = requests.get(url[num])  
    if response.status_code == 200: 
        filename1= str(i)+filename
        with open(filename1, 'wb') as f:  
            f.write(response.content)  
        print(f'图片已保存到 {filename1}')  
    else:  
        print('下载失败')
```
